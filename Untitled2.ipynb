```python
def calcular_promedio(lista):
    """Calcula el promedio."""
    return sum(lista) / len(lista) if lista else 0

def encontrar_extremos(lista):
    """Encuentra el mínimo y el máximo."""
    return (min(lista), max(lista)) if lista else (None, None)

def contar_elementos(lista):
    """Cuenta los elementos."""
    return len(lista)

def mostrar_estadisticas_completas(lista):
    """Muestra un resumen de todo."""
    if not lista:
        print("\nLa lista está vacía.")
        return

    prom = calcular_promedio(lista)
    mini, maxi = encontrar_extremos(lista)
    conteo = contar_elementos(lista)

    print("\n--- Estadísticas ---")
    print(f"Lista: {lista}")
    print(f"Conteo: {conteo}")
    print(f"Promedio: {prom}")
    print(f"Mínimo: {mini}")
    print(f"Máximo: {maxi}")

# --- 2. Menú Interactivo ---

print("Bienvenido a la Calculadora de Estadísticas")

# Pedimos la lista al usuario (separada por comas)
entrada = input("Ingresa tus números separados por comas (ej: 10, 5, 20): ")
# Convertimos la entrada en una lista de números
try:
    numeros = [float(num) for num in entrada.split(',')]
except ValueError:
    print("Error al leer la lista. Usando lista vacía.")
    numeros = []

print(f"Usando lista: {numeros}")

# Bucle principal del menú
while True:
    print("\n--- Menú ---")
    print("1. Calcular Promedio")
    print("2. Ver Mínimo y Máximo")
    print("3. Contar elementos")
    print("4. Mostrar todo")
    print("5. Salir")

    opcion = input("Elige una opción (1-5): ")

    if opcion == '1':
        print(f"Promedio: {calcular_promedio(numeros)}")

    elif opcion == '2':
        mini, maxi = encontrar_extremos(numeros)
        print(f"Mínimo: {mini}, Máximo: {maxi}")

    elif opcion == '3':
        print(f"Cantidad de elementos: {contar_elementos(numeros)}")

    elif opcion == '4':
        mostrar_estadisticas_completas(numeros)

    elif opcion == '5':
        print("¡Adiós!")
        break # Termina el bucle

    else:
        print("Opción no válida.")
```
